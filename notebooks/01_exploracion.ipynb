```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos simulados
df = pd.read_csv("../data/data_simulada.csv")

# Mostrar las primeras filas
print("Primeras filas del dataset:")
print(df.head())

# Mostrar estadísticas
print("\nEstadísticas generales:")
print(df.describe())

# Graficar temperatura por piso
plt.figure(figsize=(10,5))
for piso in df['piso'].unique():
    subset = df[df['piso']==piso]
    plt.plot(subset['timestamp'], subset['temp_C'], label=f"Piso {piso}")
plt.title("Temperatura por piso")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convertir timestamp a formato de fecha y hora
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Gráfico de temperatura por piso
plt.figure(figsize=(10,5))
for piso in df['piso'].unique():
    subset = df[df['piso'] == piso]
    plt.plot(subset['timestamp'], subset['temp_C'], label=f'Piso {piso}')
plt.title("Temperatura por piso")
plt.xlabel("Tiempo")
plt.ylabel("Temperatura (°C)")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
for piso in df['piso'].unique():
    subset = df[df['piso'] == piso]
    plt.plot(subset['timestamp'], subset['humedad_pct'], label=f'Piso {piso}')
plt.title("Humedad relativa por piso")
plt.xlabel("Tiempo")
plt.ylabel("Humedad (%)")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
for piso in df['piso'].unique():
    subset = df[df['piso'] == piso]
    plt.plot(subset['timestamp'], subset['energia_kW'], label=f'Piso {piso}')
plt.title("Consumo de energía por piso")
plt.xlabel("Tiempo")
plt.ylabel("Energía (kW)")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
